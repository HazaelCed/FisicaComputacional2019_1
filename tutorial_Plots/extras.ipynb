## Histogramas. 

`Plots.jl` nos perimite hacer histogramas de una lista dada de una manera simple utilizando la función `histogram` que tiene la siguiente sintaxis:

```julia

    historgam(datos,nbins=numero_de_cajas)

```

````julia
using Plots
gr()
````

````julia
a=rand(1000)
histogram(a,nbins=20,label="histograma 1",title="Primer histograma")
````

Podemos utilizar el argumento booleano `normed` para hacer que la frecuencia se muestre normalizada en el eje y

````julia
histogram(a,nbins=20,label="histograma 1",title="Primer histograma",normed=true)
````

## Graficación simbólica

Hasta ahora, para la mayoría de las funciones para gráficar, hemos graficado conjuntos de datos, y la mayoría de las veces estos datos se crean a aprtir de funciones ya definidas por nosotros. `Plots` nos permite ahorrarnos este paso y graficar directamente una función para un rango de valores de sus variables. Se utiliza la siguiente sintaxis:

```julia

plot(x_rango,func)

```

Podemos añadir todos los otros argumentos que nos interesan:

````julia
function f(x)
    return exp(x*log(x)) * sinh(x)
end
plot(0.01:0.01:2,f,title="Gráfica simbólica")
````

````julia
scatter(linspace(-pi+0.1,pi-0.1,201),tan,title="scatter simbólico")
````

Esta propiedad también puede aplicarse a funciones de dos variables para gráficas 3D:

````julia
function g(x,y)
    return sin(x)*exp(y*cos(y))
end
````

````julia
xs=linspace(-4*pi,pi,101)
ys=linspace(-pi,pi,101)
surface(xs,ys,g)
````

````julia
heatmap(xs,ys,g)
````
