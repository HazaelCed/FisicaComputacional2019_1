```julia
function bisección(f, a, b, ϵ=1e-5)  # 1*10^(-5)
    if f(a) * f(b) >= 0  # checar si los signos de f(a) y f(b) son iguales
         error("Para utilizar bisección, necesito que f(a) y f(b) tengan signos opuestos")
    end
    
    c = punto_medio(a, b)
    
    iterados = [c]
    
    while abs(f(c)) > ϵ
        if f(a) * f(c) < 0  # tienen signos opuestos
            a, b = a, c   # b = c  
        else
            a, b = c, b   # a = c  
        end
        
        c = punto_medio(a, b)
        push!(iterados, c)
    end
        
    return c, iterados
    
end
```

```julia
bisección(x->x^2 - 2, 0, 10, 1e-10)
```
